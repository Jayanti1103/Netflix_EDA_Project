 ### Q15:  In which year were the most shows/Movies released ?

```python
release_counts = df['release_year'].value_counts().sort_index()
plt.figure(figsize=(19,6))
release_counts.plot(kind='bar',color='red')
plt.title('Number of Netflix Titles Released Each Year')
plt.xlabel('Release Year')
plt.ylabel('Number of Titles')
plt.grid(axis='y')
plt.show()
```

```python
df['date_added'] = pd.to_datetime(df['date_added'].astype(str).str.strip(), errors='coerce')
df['year_added'] = df['date_added'].dt.year

trend = df['year_added'].value_counts().sort_index()

plt.figure(figsize=(12,6))
trend.plot(marker='o')
plt.title("Content Added to Netflix Over Time")
plt.xlabel("Year Added")
plt.ylabel("Number of Titles Added")
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(12,6))
df['country'].value_counts().head(10).plot(kind="barh", color="pink")
plt.title("Top 10 Countries with Most Netflix Content", fontsize=14)
plt.xlabel("Count")
plt.ylabel("Country")
plt.show()
```

```python
plt.figure(figsize=(8,5))
df['duration'].value_counts().head(10).plot(kind="bar", color="green")
plt.title("Most Common Durations on Netflix", fontsize=14)
plt.xlabel("Duration")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(12,6))
df['director'].value_counts().head(10).plot(kind="barh", color="orange")
plt.title("Top 10 Directors on Netflix", fontsize=14)
plt.xlabel("Count")
plt.ylabel("Director")
plt.show()
```

```python
actors = []
for a in df['cast'].dropna():
    actors.extend(a.split(', '))
    
plt.figure(figsize=(12,6))
pd.Series(actors).value_counts().head(15).plot(kind="bar", color="brown")
plt.title("Top 15 Actors on Netflix", fontsize=14)
plt.xlabel("Actor")
plt.ylabel("Count")
plt.show()
```

 ### Conclusions and Key insights

```python
''' The Netflix dataset provides valuable insights into the content library of Netflix across different countries and categories. From our analysis, we can conclude:'''

""" 1. Movies dominate over TV Shows : Netflix has a larger share of movies compared to TV shows, showing its focus on films.

 2. Content release increased significantly after 2015, reflecting Netflix’s rapid global expansion.


 3. The United States contributes the maximum number of titles, followed by India, which is one of the largest contributors of Netflix content in Asia.


 4. India has a strong presence in Movies compared to TV Shows, indicating a movie-centric entertainment culture.


 5. Genres like Dramas, Comedies, and International Movies are the most common across the platform.

 6. The dataset highlights Netflix’s shift toward regional content production, catering to country-specific audiences."""
```

```python
trend = df.groupby(['release_year','type']).size().unstack().fillna(0)

plt.figure(figsize=(12,6))
plt.stackplot(trend.index, trend['Movie'], trend['TV Show'],
              labels=['Movies','TV Shows'], alpha=0.8)
plt.legend(loc='upper left')
plt.title('Stacked Area Chart: Movies vs TV Shows Over Years')
plt.xlabel('Release Year')
plt.ylabel('Count')
plt.show()
```

```python
import seaborn as sns

top_countries = df['country'].value_counts().head(10).index
subset = df[df['country'].isin(top_countries)]

heatmap_data = pd.crosstab(subset['country'], subset['rating'])

plt.figure(figsize=(12,6))
sns.heatmap(heatmap_data, cmap='Reds', annot=False, cbar=True)
plt.title('Heatmap: Ratings Distribution Across Top Countries')
plt.xlabel('Rating')
plt.ylabel('Country')
plt.show()
```

```python
import matplotlib.pyplot as plt
type_counts = df['type'].value_counts()

plt.figure(figsize=(6,6))
plt.pie(type_counts, labels=type_counts.index, autopct='%1.1f%%', startangle=90, 
        colors=["red","blue"], wedgeprops={'width':0.4})

plt.title("Distribution of Movies vs TV Shows (Donut Chart)")
plt.show()
```

```python
trend = df.groupby(['release_year','type']).size().unstack().fillna(0)

plt.figure(figsize=(12,6))
plt.stackplot(trend.index, trend['Movie'], trend['TV Show'],
              labels=['Movies','TV Shows'], alpha=0.8, colors=["#ff9999","#66b3ff"])

plt.legend(loc='upper left')
plt.title("Movies vs TV Shows Growth Over Time (Stacked Area)")
plt.xlabel("Year")
plt.ylabel("Number of Titles")
plt.show()
```

```python
import matplotlib.pyplot as plt

top_countries = df['country'].value_counts().head(10).index
subset = df[df['country'].isin(top_countries)]

content_dist = pd.crosstab(subset['country'], subset['type'])

content_dist.plot(kind='bar', stacked=True, figsize=(12,6))
plt.title("Stacked Bar: Movies vs TV Shows in Top 10 Countries")
plt.xlabel("Country")
plt.ylabel("Count")
plt.legend(title="Content Type")
plt.show()
```
